```python
#Import Statements
import os
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
import collections
```

```python
#Importing the NLTK Stopwords
import nltk
nltk.download('stopwords')
stopwords_list = stopwords.words('english')
print(stopwords_list)
```

```python
path = 'citeseer'
filenames = os.listdir(path)   #To generate file path
```

```python
# Creating files for combination of all files from Citiseer and for storing words for each task
# Declaring Output files
combination = 'combination.txt'
a1 = 'a1.txt'
a2 = 'a2.txt'
```

### 1. Pre-Processing Steps


```python
# Combining all files from Citiseer and storing them in combination file
with open(combination, 'w') as outputfile:
  for fname in filenames:
    filepath = 'citeseer/'+fname
    with open(filepath) as inputfile:
      outputfile.write(inputfile.read()) 
    inputfile.close()
outputfile.close()
```

### Tokenization

```python
# Removing white spaces and punctuations using Tokenization
inputfile = open(combination,'r') # opening combination as read only file
text = inputfile.read() # reading combination file
lower = text.lower() # changing text to lower case
punct = re.sub(r'[^\w\s]+','',lower) # removing punctuation using regular expression
tokens = punct.split() # splitting the text to remove the white spaces
with open(a1,'w') as outputfile: # writing the tokens from previous step to a1 file
  for i in tokens:
    outputfile.write("%s " % i)
```

## 2. Frequency of words

```python
# Determining the frequency of words from tokenization
count = {} # Declaring empty dictionary 
for c in open(a1).read().split(): # reading tokens from a1 file
  if c in count:
    count[c] += 1
  else:
    count[c] = 1
print(count) # printing dictionary with frequency of words
```

 ### 2.a Total Number of Words

```python
# Total Number of words in the Collection.
file = open(a1)
data = file.read() # Reading a1 file.
words = data.split()
len(words) # getting the total number of words.
```

### 2.b Unique words

```python
# Number of unique words
len(count)
```

### 2.c Top20 Words

```python
# Sorting in decreasing order to get most frequent words first 
# Top20 words having more frequency using Lambda function
count = {a: b for a, b in sorted(count.items(), key=lambda item: item[1], reverse = True)}
top20 = {a: b for a, b in sorted(count.items(), key=lambda item: item[1], reverse = True)[:20]}
print(top20)
print(top20.keys())
```

### 2.d Stop Words from top20 words

```python
# Finding stop words from top20 words
keys = top20.keys() # Getting key names of top20 words
for i in keys:
  if i in stopwords_list:
    print(i,' ') # printing stop words in top20 words 
```

### 2.e minimum number of unique words accounting for 15% of the total number of words in the collection

```python
# minimum number of unique words accounting for 15% of the total number of words in the collection
i = 0 # Declaring variable
j = 0 # Declaring variable
min_unique = len(words)*0.15 
for p in count:
  if i < min_unique and count[p] <= (min_unique - i):
    print(p,':', count[p])
    i += count[p]
    j += 1
print('\nNumber of words:',j)
print('Total count:',i)
```
